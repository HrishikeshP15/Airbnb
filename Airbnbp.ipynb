```python
import matplotlib.pyplot as plt
import seaborn as sns

# Age distribution
plt.figure(figsize=(8,6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Count of Product Taken by Gender
gender_prod_taken = df.groupby('Gender')['ProdTaken'].mean()
print(gender_prod_taken)

# Visualization
sns.barplot(x='Gender', y='ProdTaken', data=df)
plt.title('Product Taken Rate by Gender')
plt.ylabel('Product Taken Rate')
plt.show()
```

```python
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')
```

```python
df['Gender'].value_counts()
```

```python
import pandas as pd

# KPI calculations
total_prodtaken = df['ProdTaken'].sum()
average_duration_pitch = df['DurationOfPitch'].mean()
average_monthly_income = df['MonthlyIncome'].mean()
average_pitch_satisfaction = df['PitchSatisfactionScore'].mean()

# Display the KPIs
print(f"ProdTaken: {total_prodtaken}")
print(f"Average Duration of Pitch: {average_duration_pitch}")
print(f"Average Monthly Income: {average_monthly_income}")
print(f"Pitch Satisfaction: {average_pitch_satisfaction}")
```

```python
conversion_rate = df[df['ProdTaken'] == 1].shape[0] / df.shape[0]
print(f"Customer Conversion Rate: {conversion_rate}")
```

```python
# Cross-tab for ProductPitched and Occupation
occupation_product = pd.crosstab(df['Occupation'], df['ProductPitched'])

# Plot grouped bar chart
occupation_product.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Count of CustomerID by ProductPitched and Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count of Customers')
plt.show()
```

```python
# Plot pie chart for PreferredPropertyStar
df['PreferredPropertyStar'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Count of CustomerID by PreferredPropertyStar')
plt.show()
```

```python
# Create age groups
age_groups = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65])

# Plot the count of CustomerID by age group
age_groups.value_counts().sort_index().plot(kind='bar', figsize=(10, 6))
plt.title('Count of CustomerID by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Customers')
plt.show()
```

```python
# Cross-tab for Occupation and Gender
occupation_gender = pd.crosstab(df['Occupation'], df['Gender'])

# Plot grouped bar chart
occupation_gender.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Count of CustomerID by Occupation and Gender')
plt.xlabel('Occupation')
plt.ylabel('Count of Customers')
plt.show()
```

```python
# Group by ProductPitched and sum the NumberOfTrips
trips_product = df.groupby('ProductPitched')['NumberOfTrips'].sum()

# Plot bar chart
trips_product.plot(kind='bar', figsize=(10, 6))
plt.title('Sum of Number of Trips by ProductPitched')
plt.xlabel('Product Pitched')
plt.ylabel('Sum of Number of Trips')
plt.show()
```
